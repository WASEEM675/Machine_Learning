```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
def generate_random_data(seed=0, num_samples=100):
    np.random.seed(seed)
    X = 2 * np.random.rand(num_samples, 1)
    y = 4 + 3 * X + np.random.rand(num_samples, 1)
    return X, y

# Define a function to perform linear regression
def linear_regression(X, y, learning_rate=0.01, num_iterations=1000):
    n = len(X)
    m = 0
    b = 0

    def mse(y_true, y_pred):
        return np.mean((y_true - y_pred) ** 2)

    for iteration in range(num_iterations):
        y_pred = m * X + b
        cost = mse(y, y_pred)

        dm = (-2 / n) * np.sum(X * (y - y_pred))
        db = (-2 / n) * np.sum(y - y_pred)

        m -= learning_rate * dm
        b -= learning_rate * db

        if iteration % 100 == 0:
            print(f"Iteration {iteration}: Cost = {cost}")

    return m, b

# Plot the data and regression line
def plot_data_and_regression_line(X, y, m, b):
    X_new = np.array([[0], [2]])
    y_pred = b + m * X_new

    plt.scatter(X, y, label='Original Data')
    plt.plot(X_new, y_pred, 'r', label='Regression Line')

    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    X, y = generate_random_data(seed=42)  # Change the seed as needed
    slope, intercept = linear_regression(X, y, learning_rate=0.01, num_iterations=1000)
    plot_data_and_regression_line(X, y, slope, intercept)
    print("Slope (m):", slope)
    print("Intercept (b):", intercept)
```
